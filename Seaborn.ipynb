```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') # predefined db in seaborn lib
```

```python
tips.head()
```

```python
#dist plot allows to show uni variable 
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30) # for one column only
#gives histogram
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips , kind='kde')  #for bi variate data for two variables
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm') #pairwise plot, do jointplot for every combination
#hue takes categorical column
```

```python
sns.rugplot(tips['total_bill']) #good for showing frequency in a range
```

```python
#use rugplot to explain kde - kernal density estimation plot

#normal distribution sums up together to give kde
```

```python
#Categorical plots
```

```python
#on the same df tips
```

```python
#barplot
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.mean) #aggregating data #used to compare categories

#estimator tells which aggregation function you want to add
```

```python
sns.countplot(x='sex', data=tips) #just like barplot
```

```python
#boxplot and violin plot

sns.boxplot(x='day',y='total_bill', data=tips, hue='smoker') #x= categorical , y is numerical
#categorzing based on color
```

```python
#violin plot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill',data=tips, jitter=True, hue='sex', dodge=True) #one element is categorical
#removes overlapping points
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black') #similar to striplot but better , mixture of stripplot and violinplot, swarmplot, dont use it for large datasets
```

```python
#factor plot, general form of all above plots #all plots can be created through factorplot
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
#Seaborn Matrix plots like heatmaps
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#HEATMAPS - primary way of showing matrix plot
#we need index to be values not 0,1,2,3
```

```python
tc= tips.corr()
```

```python
sns.heatmap(tc, annot=True,cmap='coolwarm') #show values in heatmpas using annot
```

```python
flights
```

```python
#convert flights in matrix
```

```python
fp=flights.pivot_table(index='month',columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=1) #optional to add linecolor and linewidths
```

```python
#second matrix plot , cluster plot
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1) #cluster months and years together based on similarity of values
#normalize results based on standard_scale to get results in a normalized way
```

```python
#Grids
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
#sns.pairplot(iris)
import matplotlib.pyplot as plt
g= sns.PairGrid(iris) #creating subplots in form of grids, it gives more control as compared to pairplot
g.map(plt.scatter)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g=sns.FacetGrid(data=tips, col='time', row='smoker') #mapping maps in the grids using facetgrid
g.map(sns.distplot, 'total_bill') 
#forms 4 plots in a grid
```

```python
g=sns.FacetGrid(data=tips, col='time', row='smoker') #mapping maps in the grids using facetgrid
g.map(plt.scatter, 'total_bill', 'tip') 
```

```python
#regression Plots for Seaborn
```

```python
#lm plot function - display linear models using seaborn
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=
          ['o','v'], scatter_kws= {'s':100}) #under the hood seaborn calls matplotlib here
```

```python
#instead of separating by colors , we can separate by columns
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day', aspect=0.6, size=8)
#to make aspect ratio better to make plots comparison better
```

```python
sns.set_style('ticks') #different styles,s
sns.countplot(x='sex', data=tips)
sns.despine(left=True) #remove top and right spines
```

```python
#styling with different method, changing  context, palettes and color
#plt.figure(figsize=(12,3))
sns.set_context('notebook', font_scale=2)
sns.countplot(x='sex', data=tips)
```
